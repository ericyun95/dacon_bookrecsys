```python
def encode_column(column):
  keys = column.unique()
  key_to_id = {key:idx for idx, key in enumerate(keys)}
  return key_to_id, np.array([key_to_id[x] for x in column]), len(keys)

def encode_df(df):
  book_ids, df['Book-ID'], num_book = encode_column(df['Book-ID'])
  user_ids, df['User-ID'], num_user = encode_column(df['User-ID'])
  return df, num_user, num_book, user_ids, book_ids
```

```python
rating_df, num_user, num_book, user_ids, book_ids = encode_df(rating_train)
print("Number of Users : ", num_user)
print("Number of Books : ", num_book)
rating_df.head()
```

```python
from scipy import sparse
```

```python
def create_embeddings(n, K):
  return 10* np.random.random((n, K)) / K

def create_sparse_matrix(df, rows, columns, column_name = 'Book-Rating'):
  return sparse.csc_matrix((df[column_name].values, (df['User-ID'].values, df['Book-ID'].values)), shape = (rows, columns))
```

```python
Y = create_sparse_matrix(rating_df, num_user, num_book)
```

```python
def predict(df, emb_user, emb_book):
  df['prediction'] = np.sum(np.multiply(emb_book[df['Book-ID']], emb_user[df['User-ID']]), axis = 1)
  return df
```

```python
def cost(df, emb_user, emb_book):
  Y = create_sparse_matrix(df, emb_user.shape[0], emb_book.shape[0])
  predicted = create_sparse_matrix(predict(df, emb_user, emb_book), emb_user.shape[0], emb_book.shape[0], 'prediction')
  return np.sqrt(np.sum((Y-predicted).power(2)/df.shape[0]))
```
